```python
nums = [2,2,1,1,1,2,2]
def majorityElement():
  n = len(nums)
  c = 0
  count = 0

  for  scanner in range(n):
    candidate = nums[scanner]


    if nums[candidate] == nums[scanner]:
      count += 1

    else:
      count -= 1

    if count == 0:
      nums[candidate] == nums[scanner]

  return candidate


majorityElement()
```
