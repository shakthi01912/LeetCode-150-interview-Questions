```python
nums = [-1,0,1,2,-1,-4]
n = len(nums)
def fjvn():
  for i in range(n):
    print("i",nums[i] )
    for j in range(i+1 , n):
      print("j",nums[j] )
      for k in range(j+1 , n):
        print("k",nums[k] )
        sum = nums[i] + nums[j] + nums[k]
        if sum == 0:

          return[nums[i] , nums[j] , nums[k]]


fjvn()
```
