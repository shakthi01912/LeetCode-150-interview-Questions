```python

S = "IV"
def romanToInt():
  tot = 0
  roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
  n = len(S)

  for i in range(n-1):
    if roman_dict[S[i] ] < roman_dict[S[i+1]] :
      tot -= roman_dict[S[i]]

    else:
      tot += roman_dict[S[i]]

  tot += roman_dict[S[n - 1]]

  return tot

romanToInt()







```
