```python
s ="0P"
cleaned = "".join(c.lower() for c in s if c.isalnum())
print(cleaned)


def vv():
  i = 0
  j = len(cleaned) - 1
  while  i < j:
    if cleaned[i] != cleaned[j]:
      return False

    else:
      i+=1
      j-=1
  return True

vv()
```

```python
s = "abc"
t = "ahbgdc"
def isSubsequence():
  i, j = 0, 0

  while i < len(s):
    if s[i]  != t[j]:
      j += 1

    i += 1

  return i == len(s)

isSubsequence()
```
