```python
arr = [7 , 10 , 4 , 3, 20 , 15]
n = len(arr)
for a in range(1,n):
  print("outer loop: " , a)
  for i in range(n-a):
    print("   inner loop: " ,i )
    if arr[i] > arr[i+1]:
      print("       inner loop  checking: " ,arr[i] , arr[i+1] )
      temp = arr[i] #temp
      print("             storing i's vale" , temp)
      arr[i] = arr[i+1]
      print("             inner loop i value is changed with adjacent",arr[i+1] ,"value: " , arr[i])
      arr[i+1] = temp
      print("             adjacent is changed with value: " , arr[i+1])

    else:
      print("       no need swap between" , arr[i] , arr[i+1])
  print(" ")



print(arr)
```
