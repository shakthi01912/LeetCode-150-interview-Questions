```python
arr1 = [1, 4, 5, 7]
m = len(arr1)
arr2 = [10, 20, 30, 40]
n = len(arr2)
x = 32

def smallestAbs():
  min= float('inf')
  i  = 0
  j =  n-1
  while i < m and  j >= 0:
    sum = arr1[i] + arr2[j]
    abss = abs(sum - x)
    if sum > x:
      j -= 1
    else:
      i += 1

    if min >  abss:
      min = abss
      minI = i
      minJ = j
      return [minI , minJ]

smallestAbs( )
```
