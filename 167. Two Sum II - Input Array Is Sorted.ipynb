```python
numbers = [2,7,11,15]
target = 9

n = len(numbers)



def twoSum():
  i = 0
  j = n - 1
  temp = []
  while i < j :
    sum = numbers[i] + numbers[j]

    if sum == target:
      temp.append(i+1)
      temp.append(j+1)

      return temp

    elif sum > target:
      j -= 1

    else:
      i += 1







twoSum()
```
