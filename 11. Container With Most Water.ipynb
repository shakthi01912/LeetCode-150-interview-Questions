```python
height = [1,2 ,1]

def maxArea():
  n = len(height)
  i = 0
  j = n - 1
  s = 0
  sum = 0
  maxs = 0

  while i < j:
    if  height[i] > height[j]:
      s = height[j]
      sum = s * (j-i)
      j -= 1

    else:
      s = height[i]
      sum = s * (j-i)
      i += 1


    if maxs < sum:
      maxs = sum

  return maxs








maxArea()
```
