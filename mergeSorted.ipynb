```python
count = 0
def mergeSort(arr):
    global count
    count += 1

    if len(arr) <= 1:
        print("------------------base condition is matched -------------------", arr)
        return arr

    print("mergeSort logic")
    #print("array length is: " , len(arr))
    mid = len(arr) // 2
    #print("mid is: " , mid )
    leftHalf = arr[:mid]
    print(" left half: " , leftHalf)
    rightHalf = arr[mid:]
    print(" right half: " , rightHalf)

    print("----------------------------------------------------------------------function",  count)
    print(" ")


    sortedLeft = mergeSort(leftHalf)
    print("       sub left : " , sortedLeft)

    sortedRight = mergeSort(rightHalf)
    print("       sub right : " , sortedRight)

    print ("        ......release left: " , sortedLeft)
    print ("        ......release right: ", sortedRight)
    return merge(sortedLeft, sortedRight)

def merge(left, right):
    print(" ")
    print("       >>>>>>>>unwind  left: " , left , "right: " , right)
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            print("       comparison")
            result.append(left[i])
            print("       added to the list" ,left[i] )
            i += 1
            print("           here")
        else:
            result.append(right[j])
            print("       added to the list" ,  right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

unsortedArr = [3, 7, 6, -10, 15, 23.5]
sortedArr = mergeSort(unsortedArr)
print("Sorted array:", sortedArr)
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

firstP = m -1
secondP= n-1
listP = (m+n) - 1

while  firstP >= 0 and  secondP >= 0 :
  if nums1[firstP]  > nums2[secondP]:
    nums1[listP] = nums1[firstP]
    firstP -= 1
    listP -= 1

  else:
    nums1[listP] = nums2[secondP]
    secondP -= 1
    listP -= 1

while secondP >= 0:
    nums1[listP] = nums2[secondP]
    secondP -= 1
    listP -= 1


print(nums1)

```

```python
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

firstP = m - 1  # Pointer to the last valid element in nums1
secondP = n - 1  # Pointer to the last element in nums2
listP = (m + n) - 1  # Pointer to the last position in nums1

while secondP >= 0:  # Merge until all elements of nums2 are processed
    if firstP >= 0 and nums1[firstP] > nums2[secondP]:  # Compare elements if nums1 has valid values
        nums1[listP] = nums1[firstP]
        firstP -= 1
    else:  # If nums2 element is larger or nums1 is exhausted
        nums1[listP] = nums2[secondP]
        secondP -= 1
    listP -= 1  # Move to the next position

print(nums1)
```
